```python
from pandas_profiling import ProfileReport
```

```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
df_wine = load_wine(as_frame=True).data
df_wine_target = load_wine(as_frame=True).target
```

```python
df = pd.concat([df_wine, df_wine_target], axis=1)
df.head()
```

```python
profile = ProfileReport(df, title="Wine Report")
```

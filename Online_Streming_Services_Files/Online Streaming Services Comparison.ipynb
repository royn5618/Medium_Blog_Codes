```python
def reset_bar_width(ax, new_width):
    for patch in ax.patches:
        old_width = patch.get_width()
        # print(current_width)
        patch.set_width(new_width)
        shift = old_width - new_width
        patch.set_x(patch.get_x() + shift * .5)
```

```python
fig = plt.figure(figsize=[6,6])
sns.set_style('darkgrid')
ax = sns.countplot(x="Age", 
                   data=df, order=['7+', '13+', '16+', '18+', 'all'], 
                   palette=sns.cubehelix_palette(8, reverse=True))
reset_bar_width(ax, 0.3)
plt.ylabel("Number of TV Shows ")
plt.show()
```

## Text Analysis of the Names on the TV Shows

My Prediction here is - Love and life are two words which are going to be among the top five in the wordcloud of 100 words.

```python
import nltk
# nltk.download('punkt')
from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import PIL.ImageOps
from wordcloud import ImageColorGenerator
```

```python
df['Title Tokens'] = df['Title Cleaned'].apply(lambda x: word_tokenize(x))
```

```python
df['Title Tokens']
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')

lm = WordNetLemmatizer()
df['Title Lemmas'] = df['Title Tokens'].apply(lambda x: [lm.lemmatize(token) for token in x])
df['Title Lemmas'].head(5)
```

```python
token_list = [str(each_token) for each_list in list(df['Title Lemmas']) for each_token in each_list]
len(token_list)
```

It also caught my eye that man and men are difficult convertions for the current lemmatization methods. So, I will manually change that by str replace.

```python
token_list = ['man' if token == 'men' else token for token in token_list]
len(token_list)
```

```python
# mask_image = 'blue_circle.png'
mask_image = resources_folder + 'ert.jpg'
image_mask = np.array(Image.open(mask_image))
image_colors = ImageColorGenerator(image_mask)
    
wordcloud = WordCloud(width = 2000, height = 1200, max_words = 500, normalize_plurals = True, colormap='YlOrBr', 
                      mask=image_mask,
                      background_color = 'white', font_step = 1, relative_scaling = 0.8,
                      collocations = False, include_numbers = False).generate(' '.join([each_token.upper() for each_token in token_list]))                  
plt.figure(figsize = (20,13))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig(resources_folder + 'demo7.png', transparent=True)
plt.show()
```

```python
all_words_list = list(wordcloud.words_.keys())
```

```python
word_freq_dict = {}
for each_word in all_words_list:
    count = token_list.count(each_word.lower())
    word_freq_dict[each_word] = count
word_freq_dict = dict(sorted(word_freq_dict.items(), key=lambda kv: kv[1], reverse=True)[:10])
word_freq_dict
```

```python
fig = plt.figure(figsize=[15,12])
sns.set_style('dark')

ax = sns.barplot(x=list(word_freq_dict.keys()), y=list(word_freq_dict.values()),
            palette = ['#404040'] * 10
           )
for i, xlabel_text in enumerate(list(word_freq_dict.keys())):
    ax.text(i - .095 , 5, xlabel_text, color='#FFC000', fontweight='bold', rotation=90, fontsize=20)
    if i == 0:
        ax.text(i - .095 , 93, word_freq_dict[xlabel_text], color='#FFC000', fontweight='bold', rotation=90, fontsize=20)
    else:
        ax.text(i - .095 , 95-(list(word_freq_dict.values())[0] - list(word_freq_dict.values())[i]), word_freq_dict[xlabel_text], color='#FFC000', fontweight='bold', rotation=90, fontsize=20)
        
plt.yticks([])
plt.xticks([])
reset_bar_width(ax, 0.5)
plt.savefig(resources_folder + 'demo8.png', transparent=True)
```
